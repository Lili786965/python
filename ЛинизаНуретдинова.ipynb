1. Дана последовательность слов. Напечатать те слова последовательности,
которые отличны от последнего слова и удовлетворяют следующему свойству:
слово симметрично

```python
def symmetric_words(words):
    if not words:
        return
    last_word = words[-1]
    symmetric_words = []
    for word in words[:-1]:
        if word != last_word and word == word[::-1]:
            symmetric_words.append(word)
    for word in symmetric_words:
        print(word)
sequence = ["level", "водоводов", "hello", "шалаш", "python", "laptop"]
symmetric_words(sequence)
```

2. В символьной строке заменить все двоеточия (:) точкой с запятой (;).
Подсчитать количество замен

```python
def replace_colons(text):
    replaced_text = text.replace(':', ';')
    count = text.count(':')
    return replaced_text, count

input_string = "Change the :: to ;;"
new_string, replacements = replace_colons(input_string)

print(input_string)
print("After change:", new_string)
print("Number of replacements:", replacements)
```

3. Программа. Дана непустая последовательность непустых слов из латинских букв;
соседние слова отделены друг от друга запятой, за последним словом—точка.
Определить количество слов, которые начинаются с буквы а;

```python
def count_a(sequence):
    words = sequence[:-1].split(',')
    count = sum(1 for word in words if word.strip().lower().startswith('a'))
    return count
input_sequence = "apple,april,angry,orange,Andrew,cabage."
result = count_a(input_sequence)
print("Number of 'a' words:", result)
```

1. Дан символьный файл, содержащий по крайней мере один символ пробела.
Удалить все его элементы, расположенные после первого символа пробела,
включая и этот пробел.


```python
!pip install faker
from faker import Faker
import random

fake = Faker()

def generate_random_text():
    return fake.text(max_nb_chars=random.randint(50, 200))

def create_file(file_name, num_lines=10):
    with open(file_name, 'w', encoding='utf-8') as file:
        for _ in range(num_lines):
            file.write(generate_random_text() + '\n')

def delete_after_space():
    input_filename = "input_task1.txt"
    create_file(input_filename, num_lines=5)

    with open(input_filename, 'r', encoding='utf-8') as file:
        content = file.read()
    print(content)
    print("\n" + "="*50 + "\n")

    first_space_index = content.find(' ')
    if first_space_index != -1:
        result = content[:first_space_index]
    else:
        result = content
    print(result)

delete_after_space()
```

2. Дан файл вещественных чисел. Заменить в нем все элементы на их квадраты.

```python
from faker import Faker
import random

fake = Faker()

def turnToSquare():
    input_filename = "input_task2.txt"
    with open(input_filename, 'w', encoding='utf-8') as file:
        for _ in range(10):
            file.write(f"{random.uniform(1, 10):.2f}\n")

    with open(input_filename, 'r', encoding='utf-8') as file:
        numbers = [float(line.strip()) for line in file if line.strip()]

    squared_numbers = [num ** 2 for num in numbers]
    for original, squared in zip(numbers, squared_numbers):
        print(f"{original:<10.2f} | {squared:<10.2f}")

turnToSquare()
```

3. Дан файл целых чисел, содержащий не менее четырех элементов. Вывести
первый, второй, предпоследний и последний элементы данного файла

```python
import random

def printing():
    numbers = [random.randint(1, 100) for _ in range(random.randint(4, 10))]

    print(numbers)
    if len(numbers) < 4:
        print("Ошибка: последовательность содержит меньше 4 чисел!")
        return
    print(f"{'1':<15} | {numbers[0]:<10}")
    print(f"{'2':<15} | {numbers[1]:<10}")
    print(f"{'prefinal':<15} | {numbers[-2]:<10}")
    print(f"{'last':<15} | {numbers[-1]:<10}")

printing()
```
