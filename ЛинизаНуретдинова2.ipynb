Создайте класс Client с полями имя, идентификационный номер и
баланс. Реализуйте методы update_balance(amount), который обновляет
баланс, и display_info(), который выводит информацию о клиенте

Определите класс Account, который наследуется от класса Client.
Класс Account должен добавлять методы для пополнения счёта, списания
средств и отображения информации о счёте.

```python
class Client:
    def __init__(self, name, number, balance):
        self.client_name = name
        self.account_number = number
        self.balance = balance

    def update_balance(self, amount):
        self.balance += amount
        print(f"Зачислено {amount}")

    def display_info(self):
        print(f"Имя: {self.client_name} \nНомер аккаунта: {self.account_number} \nСумма баланса: {self.balance}")

class Account(Client):
    def add(self, amount):
        self.balance += amount
        print(f"На счет зачислено: {amount}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Со счета снято: {amount}")
        else:
            print("Недостаточно средств на счете")

    def display(self):
        print(f"Номер счета: {self.account_number} \nСумма баланса: {self.balance}")

acc1 = Client("Clint", "123456577", 120)
acc1.update_balance(1000)
acc1.display_info()

acc2 = Account("Alice", "7654321", 500)
acc2.add(500)
acc2.withdraw(300)
acc2.display()
```

Создать класс Account, в котором поля accountNumber и balance
будут приватными. Реализовать методы для безопасного пополнения и
списания средств, а также получения информации о счёте.


```python
class Account:
    def __init__(self, account_number, balance):
        self.__account_number = account_number
        self.__account_balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__account_balance += amount
            print(f"Зачислено {amount}. Новый баланс: {self.__account_balance}")
        else:
            print("Сумма для зачисления должна быть положительной.")

    def withdraw(self, amount):
        if amount > 0:
            if self.__account_balance >= amount:
                self.__account_balance -= amount
                print(f"Снято {amount}. Новый баланс: {self.__account_balance}")
            else:
                print("Недостаточно средств на счете.")
        else:
            print("Сумма для снятия должна быть положительной.")

    def get_account_info(self):
        print(f"Номер счета: {self.__account_number}")
        print(f"Баланс: {self.__account_balance}")

account = Account("123456789", 1000.0)
account.get_account_info()
account.deposit(500)
account.withdraw(200)
account.get_account_info()
```

Путешествие и достопримечательность. Создайте класс
`Путешествие`, в котором есть достопримечательности, которые посетят
путешественники. У класса `Достопримечательность` есть поля `название` и

```python
class Sights:
    def __init__(self,name, description):
        self.name = name
        self.description = description

    def show_info(self):
        print(f"Sight: {self.name}")
        print(f"Description: {self.description}")

class Travel:
    def __init__(self, name):
        self.name = name
        self.sights = []

    def add_sights(self, sights):
        self.sights.append(sights)
        print(f"Sight '{sights.name}' is added to '{self.name}'.")

    def show_plan(self):
        print(f"Trip: {self.name}")
        for sights in self.sights:
            sights.show_info()

sight1 = Sights("Big Ben", "in London")
sight2 = Sights("The Buckingham Palace", "is the residence of English royalty")

traveling = Travel("Trip to the UK")
traveling.add_sights(sight1)
traveling.add_sights(sight2)
traveling.show_plan()
```

Создать абстрактный класс Bank с абстрактным методом
calculateInterest(). Реализовать подклассы SavingsBank и InvestmentBank,
которые реализуют метод calculateInterest().

```python
from abc import ABC, abstractmethod

class Bank(ABC):
    def __init__(self, sum):
        self.sum = sum

    @abstractmethod
    def calculate_interest(self):
        pass

class SavingsBank(Bank):
    def __init__(self, sum, interest_rate):
        self.sum = sum
        self.interest_rate = interest_rate

    def calculate_interest(self):
        interest = self.interest_rate * self.sum
        print(f"Savings Percentage: {interest}")

class InvestmentBank(Bank):
    def __init__(self, sum, perc):
        self.sum = sum
        self.perc = perc

    def calculate_interest(self):
        interest = self.sum * self.perc
        print(f"Investment percentage: {interest}")

savings = SavingsBank(1000, 0.03)
savings.calculate_interest()

investment = InvestmentBank(1000, 0.05)
investment.calculate_interest()
```

Определите метод transferFunds() в классе Account, который будет
переводить средства между счетами. Создайте подклассы SavingsAccount и
CheckingAccount, которые будут переопределять этот метод для выполнения
специфических типов переводов

```python
from abc import ABC, abstractmethod

class Account(ABC):
    @abstractmethod
    def transferFunds(self, transfer_acc, sum):
        pass

class SavingsAccount(Account):
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def transferFunds(self, transfer_acc, sum):
        if sum <= 0:
            print("Сумма должна быть положительной")
            return
        if self.balance >= sum:
            self.balance -= sum
            transfer_acc.balance += sum
            print(f"Переведено {sum} со счета {self.account_number} на счет {transfer_acc.account_number}.")
        else:
            print("Недостаточно средств для перевода.")

class CheckingAccount(Account):
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def transferFunds(self, transfer_acc, sum):
        if sum <= 0:
            print("Сумма должна быть положительной")
            return
        if self.balance >= sum:
            self.balance -= sum
            transfer_acc.balance += sum
            print(f"Переведено {sum} со счета {self.account_number} на счет {transfer_acc.account_number}.")
        else:
            print("Недостаточно средств для перевода.")

savings1 = SavingsAccount("0123", 1000)
savings2 = SavingsAccount("0124", 500)
checking1 = CheckingAccount("123", 1500)
checking2 = CheckingAccount("124", 2000)

savings1.transferFunds(savings2, 200)
checking1.transferFunds(checking2, 300)
savings1.transferFunds(checking1, 100)
checking1.transferFunds(savings1, 100)
```
